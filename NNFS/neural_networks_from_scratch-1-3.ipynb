```python
import numpy as np
```

```python
# <NEURON>

# every neuron only has one bias
# every neuron has number of weights == number of inputs
# output: weights DOT inputs PLUS bias
# DOT: multiplying same index of one array in other array
# in e.g. below: (1.2*3.1 + 5.1*2.1 + 2.6*8.7) + 3


inputs = np.array([1.2,5.1,2.6]) # these are outputs of input neurons
weights = np.array([3.1,2.1,8.7])
bias = 3

output = np.dot(weights,inputs) + bias
print(output)

# </NEURON>
```

```python
# 4 input neurons and 3 output neurons
# inputs are 4 for 1 neuron so 4x1 input matrix
# so for each output neuron, it will have 4 values of weights and 1 bias
# so for 3 output neuron, it will be an 3x4 matrix of weights
# since 1 neuron has 1 bias, total 3 neurons: 3x1 matrix of bias
# we need both weights and bias, bias offsets the output 

inputs = np.array([1,2,3,2.5])
weights = np.array([    [0.2,0.8,-0.5,1.0],
                        [0.5,-0.91,0.26,-0.5],
                        [-0.26,-0.27,0.17,0.87]    ])
biases = np.array([2,3,0.5])


# for dot product, no. of cols in first matrix == no. of rows in second matrix
# and the result will be of shape: rows in first matrix X cols in second matrix 
# that is 3x1
# here 3x4 and 4x1 so we'll get a matrix of 3x1
# adding to bias matrix of 3x1
# we get output matrix of 3x1

print(weights,weights.shape)
print(biases, biases.shape)


outputs =  np.dot(weights,inputs) + biases
print(outputs) 
```
